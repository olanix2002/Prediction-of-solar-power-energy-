```python
x = data.drop('generated_power_kw', axis=1)
y = data['generated_power_kw']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
print("x_train: ", x_train.shape)
print("x_test: ", x_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from lazypredict.Supervised import LazyRegressor

reg = LazyRegressor(verbose=0,ignore_warnings=False, custom_metric=None )
models,predictions = reg.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
model1 = RandomForestRegressor()
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2', None]
}

# Set up the RandomizedSearchCV object
rs = RandomizedSearchCV(model1, param_grid, cv=5, n_iter=10, scoring='neg_mean_absolute_error')

# Fit the model
rs.fit(x_train, y_train)

# Print the best parameters and score
print(rs.best_params_)
print(rs.best_score_)

# Use the best model to make predictions on the test data
y_pred = rs.predict(x_test)
#In this example, the RandomizedSearchCV object is initialized with the random forest model and a parameter grid that specifies the values to be searched for each hyperparameter. The fit method trains the model and performs the hyperparameter tuning, and the best model and its hyperparameters are stored in the best_params_ attribute. You can then use the best model to make predictions on the test data.

#You can also use the GridSearchCV class, which performs an exhaustive search over the specified parameter grid rather than a random search. The syntax is similar to RandomizedSearchCV, except that you don't need to specify the n_iter parameter.

#It's important to note that hyperparameter tuning can be a time-consuming process, especially for large datasets or complex models. It's also important to choose an appropriate scoring metric and to validate the model using a separate test set or using cross-validation to avoid overfitting.
```

```python
from sklearn.feature_selection import RFECV

model_1 = RandomForestRegressor(n_estimators=400, min_samples_split=3, min_samples_leaf=3, max_features='auto', max_depth=100)
rfecv = RFECV(model_1 , step = 1, min_features_to_select = 7 , n_jobs = -1)
rfecv.fit(x_train , y_train)
```

```python
rfecv.support_
```

```python
rfecv.ranking_
```

```python
selected_features = np.where(rfecv.support_)[0]
x_train = x_train.iloc[:,selected_features]
x_test = x_test.iloc[:,selected_features]
```

```python
model_1.fit(x_train , y_train)
```

```python
model_1.score(x_test , y_test)
```
